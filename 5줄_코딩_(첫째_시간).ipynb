#데이터의 통계를 내 봅니다.

```python
data.describe()
```

count는 데이터의 개수,</br>
mean은 평균, </br>
std는 표준편차, </br>
min은 최소값, </br>
25%는 1번째 quartile (25% 순위에 해당하는 값),</br>
50%는 2번째 quartile (중앙값),</br>
75%는 3번째 quartile (75% 순위에 해당하는 값),</br>
max 는 최대값을 말합니다.</br>
